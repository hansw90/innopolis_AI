- try except else

```
n = [10, 20, '3']
```

```
try :
    res = map(div, n)
    print(list(res))
except TypeError :
    print('자료형을 확인하세요. int만 사용이 가능합니다.')
except ZeroDivisionError :
    print('0으로 나누면 에러가 납니다.')
else :
    print('정상적으로 실행했습니다.')
```

- try except finally

```
n = [10, 20, 0]
```

```
try :
    res = map(div, n)
    print(list(res))
except TypeError :
    print('자료형을 확인하세요. int만 사용이 가능합니다.')
except ZeroDivisionError :
    print('0으로 나누면 에러가 납니다.')
else :
    print('정상적으로 실행했습니다.')
finally :
    print("코드의 실행이 끝났습니다.")
```

### quiz 3
nickname은 문자로 된 입력값을 받으면 화면에 닉네임을 출력해주는 함수입니다.\
그러나 닉네임은 고유의 값으로 이미 존재하는 닉네임을 설정할 수는 없습니다.\
따라서 현재 존재하는 닉네임 값을 입력으로 받으면 에러가 나야합니다.(현재 존재하는 닉네임은 "MJ"와 "YJ"라고 가정해봅시다.) 즉, "MJ"나 "YJ"라는 닉네임을 input으로 받게 되면 에러가 발생해야합니다.\
에러가 나면 "닉네임이 이미 존재합니다."라는 에러 메세지를 화면에 출력할 수 있도록 에러를 만들어 봅시다.

```
# 정답을 적어주세요
nicknames = [ "MJ", "YJ"]


class nicknameError(Exception) :
  def __init__(self) :
    super().__init__("닉네임이 이미 존재합니다.")


def nickname(nick) :

  if nick in nicknames :
    raise nicknameError
  else :
    print(nick)
```

```
nickname("H")
```

```
nickname("MJ")
```
