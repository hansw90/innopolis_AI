## ufunc 유니버셜 함수 

**유니버셜 함수**  
ndarray 안에 있는 데이터 원소별로 연산을 수행하는 함수  
하나 이상의 스칼라 값을 받아서 하나 이상의 스칼라 값을 반환하는 간단한 함수를 **고속**으로 수행할 수 있는 벡터화 된 함수

|Function|설명|
|---|---| 
|abs, fabs|각 원소의 절대값. 복소수가 아닌 경우 fabs를 쓰면 빠른 연산이 가능|
|sqrt|각 원소의 제곱근 계산(arr**0.5)|
|square|각 원소의 제곱을 계산(arr**2)|
|exp|각 원소에 지수e^x 계산|
|log, log10, log2, log1p|자연로그, 밑10인 로그, 밑2인 로그, log(1+x)|
|sign|각 원소의 부호(양수 1, 영 0, 음수 -1)|
|ceil|각 원소의 값보다 같거나 큰 정수 중 가장 작은 값|
|floor|각 원소의 값보다 같거나 작은 정수 중 가장 큰 값|
|rint|각 원소의 소수자리를 반올림|
|modf|각 원소의 몫과 나머지를 각각의 배열로 반환|
|isnan|각 원소가 숫자인지 아닌지. 불리언 배열로 반환|
|isfinite, isinf|각 원소가 유한한지, 무한한지. 불리언 배열로 반환|
|cos, cosh, sin, sinh, tan, tanh|일반 삼각함수, 쌍곡삼각함수|
|arccos, arccosh, arcsin, arcsinh, arctan, arctanh|역삼각함수|





````python
arr = np.arange(0, 5, 0.5).reshape(2, -1)
arr
````

````python
np.exp(arr)
````

````python
arr[:, 0] = np.inf
np.isinf(arr)
````

## 집합 관련 함수

np.unique(): 중복된 원소를 제거하고 남은 원소를 정렬된 형태로 반환

````python
arr = np.array(['Kim', 'Park', 'Kim', 'Lee', 'Kim', 'Na', 'Kim', 'Park', 'Lee', 'Na', 'Bae'])
np.unique(arr)
````

np.intersect1d(x, y): 배열 x, y에 공통으로 들어있는 원소를 정렬해서 반환

````python
arr1 = np.array(['a', 'b', 'c', 'd'])
arr2 = np.array(['e', 'd'])
np.intersect1d(arr1, arr2)
````

np.union1d(x, y): 배열 x, y의 합집합을 반환

````python
np.union1d(arr1, arr2)
````

np.in1d(x, y): x의 원소가 y의 원소에 포함 되는지 나타나는 불리언 배열 반환

````python
np.in1d(arr1, arr2)
````

````python
np.in1d(arr2, arr1)
````

np.setdiff1d(x, y): x와 y의 차집합

````python
np.setdiff1d(arr1, arr2)
````

````python
np.setdiff1d(arr2, arr1)
````

setxor1d(x, y): 한 배열에는 포함되지만 두 배열 모두에는 포함되지 않는 원소들의 집합 반환

````python
np.setxor1d(arr1, arr2)
````

## 선형대수

행렬의 곱셉, 행렬식, 정사각 행렬 수학 등 다양한 선형 대수 관련 함수가 있음

### np.dot(x, y): x와 y의 dot product  
    - x, y가 1차원이 아니고 매트릭스나 상수일 때도 연산을 해주지만 권장하지는 않음.
    - x, y가 둘다 2차원인 경우 np.matmul 또는 x@y 권장
    - x, y 둘 중 하나라도 0차원인 경우(scalar) np.multiply 또는 x*y 권장


````python
x = np.random.randint(-5, 5, size=10)
y = np.random.randint(-10, 10, size=10)
x.shape, y.shape
````

````python
x.dot(y)
````

````python
np.dot(x, y)
````

````python
x = np.array([[1, 2, 3], [4, 5, 6]])
y = np.array([[2, -3], [1, 6], [-1, 2]])
x.shape, y.shape
````

````python
np.dot(x, y)  # x와 y가 둘 다 1차원인 경우만 권장. 아닌 경우는 권장하지 않음. 
````

    - 차원이 다를 때 broadcast 일어남.


````python
a = np.random.randint(-3, 3, 10).reshape(2, 5)
b = np.random.randint(0, 5, 15).reshape(5, 3)
a.shape, b.shape
````

````python
np.matmul(a, b)
````

````python
np.matmul(b, a)  # matrix곱할때는 shape이 맞도록
````

### np.linalg.eig(x)
    - 매트릭스의 고유값, 고유 벡터를 반환
    - returns : w-고유값, v-고유벡터
    - 고유벡터 v[:, i] 에 대응하는 고유값은 w[i].

````python
matrix = np.array([[1, 2], [4, 6]])
````

````python
w, v = np.linalg.eig(matrix)
````

````python
v # eigenvalues
````

````python
w  # normalized (unit “length”) eigenvectors
````

````python
np.matmul(matrix, v[:, 0])
````

````python
w[0]*v[:, 0]
````

### np.linalg.inv(x)
    - matrix의 inverse

````python
matrix = np.array([[10, 20], [30, 40]])
````

````python
inv_matrix = np.linalg.inv(matrix)
inv_matrix
````

````python
matrix@inv_matrix
````

````python
# inverse가 존재하지 않는 경우 에러 발생
matrix = np.array([[10, 20], [-1, -2]])
np.linalg.inv(matrix) 
````

### np.linalg.solve(A, b)
    - Ax = b 인 해 x를 반환

````python
A = np.array([[3, 1, -2], [0, -5, 7], [4, 0, 1]])
b = np.array([-5,  1, -1])
x = np.linalg.solve(A, b)
x
````

````python
np.matmul(A, x), b
````

````python
# 해가 존재하지 않는 경우 에러 발생
A = np.array([[1, -2, 3], [2, 9, 6], [-1, 7, -3]])
b = np.array([1,  2, 3])
x = np.linalg.solve(A, b)
x
````
