### 클래스 기본

```python
class MyClass() :
    class_var = '클래스 변수'
    
    @classmethod
    def class_method(cls):
        print('클래스의 메소드')
    
    def instance_method(self):
        self.instance_var = '인스터스 변수'
        print("인스턴스의 메소드")
```

```python
MyClass.class_method()
```

```python
MyClass.instance_method()
```

```python
cls = MyClass()
```

```python
cls.instance_method()
```

### instance vs. class

```python
class Room() :
    
    def add_person(self, name):
        self.room = []
        self.room.append(name)
```

```python
r1 = Room()
r2 = Room()

r1.add_person('뽀로로')
r2.add_person('펭수')
```

```python
r1.room
```

```python
r2.room
```

```python
Room.c_room
```

```python
class Room2() :
    
    c_room = []

    def __init__(self) :
        self.c_room = []
    
    @classmethod
    def add_c_room(cls, name):
        cls.c_room.append(name)

    def add_person(self, name):
        self.room.append(name)
```

```python
r3 = Room2()
r4 = Room2()

r3.add_c_room('뽀로로')
r4.add_c_room('펭수')
```

```python
Room2.c_room
```

```python
# 인스턴스
r3.c_room
```

```python
r3.room
```

```python
r3.add_person('홍길동')
r4.add_person('뚝딱이')
```

```python
r3.room
```

```python
r4.room
```

```python
Room2.c_room
```

+ 클래스의 상속

```python
class MyPhone():
    def __init__(self, model, color) :
        self.model = model
        self.color = color
    
    def set_name(self, name):
        self.user = name
        print("사용자의 이름은 : %s" %self.user)
    
    def set_number(self, number):
        self.number = number
```

```python
class MyPhone2(MyPhone):
    def has_case(self, val=False) :
        self.case = val
```

```python
p2 = MyPhone2('갤럭시S8', 'black')
```

```python
p2.set_name("MJ")
```

```python
p2.has_case(True) 
```

```python
p2.has_case
```

+ 오버라이딩
