## 라이브러리 requests를 이용하여 json 파일을 가져오고, 원하는 데이터를 추출해보기


```
import requests
```

```
import pandas as pd
```

 ###  requests.get('사이트주소') 
웹 페이지 정보 가져오기  


```
response= requests.get('https://www.naver.com/srchrank?frm=main')  # 
type(response)
```

#### Response.json()
json 형식으로 반환

```
# .json 형태로 불러올수 있는 사이트는 아래와 같이 불러올 수 있다.
print(type(response.json()))
```

```
j = response.json()
j
```

```
j.keys()
```

```
ranking = j['data']
ranking
```

```
for r in ranking:
    rank = r.get('rank')
    keyword = r.get('keyword')
    print('{:2}위 {}'.format(rank, keyword))
```

```
# pandas 형식으로 바꿔보기 
df = pd.DataFrame(ranking)
```

```
df
```

```
# 전체 코드
response= requests.get('https://www.naver.com/srchrank?frm=main')
j = response.json()
ranking = j['data']
df = pd.DataFrame(ranking)
df
```
