####문제 3
고객의 개인정보보호를 위하여 이름을 비식별화 하여 출력하기   
예시) 홍길동 → 홍\*동

```python
names = ['홍길동', '홍계월', '김철수', '이영희', '박첨지']
```

```python
# 정답을 써주세요

for i,name in enumerate(names) : 
  k = []
  for j,word in enumerate(name) :
    k.append(word)
    
  k[1] = "*"
  print(''.join(k))
```

```python
for m in names:
  a = m.replace(m[1], '*')
  print(a)
```
